````python
df = pd.DataFrame(clean['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
clean['STATE'] = list(map(clean_state, clean['STATE']))
````

````python
# This looks a bit dodgy, a mix of strings and values. I'll replace the strings 
clean['NOEXCH'].value_counts()
````

````python
# Yay
clean['NOEXCH'] = clean['NOEXCH'].replace("0", 0)
clean['NOEXCH'] = clean['NOEXCH'].replace("1", 1)
````

````python
# Need to remove the NaNs and replace them with the median

mode_values = clean.mode().iloc[0]  # Get the first row of the mode DataFrame

# Replace NaN values with the mode for each column
clean_full = clean.fillna(mode_values)
````

````python
nums = data.select_dtypes(np.number)
nums
````

````python
# Here's my look at how many contain empty values
df = pd.DataFrame(nums.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
````

````python
# 91 columns is far too many to go through one by one, so I will replace all the none values with the mean 

def replace_null_with_mean(df):
    for column in df.columns:
        mean_value = df[column].mean()
        df[column].fillna(mean_value, inplace=True)
    return df

# Example usage:
cleaned_nums = replace_null_with_mean(nums)
cleaned_nums
````

````python
# Now I want to reduce the number of columns, so will look at the variance 
````

````python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
numerical_scaled = scaler.fit_transform(cleaned_nums)

# Convert the scaled data back to a DataFrame with column names
numerical_scaled = pd.DataFrame(numerical_scaled, columns=cleaned_nums.columns)
numerical_scaled
````

````python
from sklearn.feature_selection import VarianceThreshold 

var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))

# Fit the VarianceThreshold selector
sel.fit(numerical_scaled)

# Get the indices of the selected features
selected_indices = sel.get_support(indices=True)

# Get the selected column names
selected_columns = numerical_scaled.columns[selected_indices]

# Transform the original DataFrame and preserve the column names
temp = pd.DataFrame(sel.transform(numerical_scaled), columns=selected_columns)
temp
````

````python
targets = pd.read_csv('target.csv') 

X = temp
y = targets['TARGET_B']

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# or from sklearn.feature_selection
kbest = SelectKBest(chi2, k=10).fit_transform(X, y)
# Here we choose 10 so that is easier to analyze results later, as we will see
selected = pd.DataFrame(kbest)
model = SelectKBest(chi2, k=10).fit(X, y)
df = pd.DataFrame(data = model.scores_, columns = ['score'])
df['Column'] = temp.columns
list = (df.sort_values(by = ['score'], ascending = False).head(10))
````
