````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
````

````python
data = pd.read_csv("learningSet.csv")
````

````python
print(data.shape)
````

````python
cats = data.select_dtypes(object)
cats
````

````python
# Checking the empty values in each column
# Looks like only one row has null values in it - seems a bit weird 
nulls_percent_df = pd.DataFrame(cats.isna().sum()/len(cats)).reset_index()
nulls_percent_df
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df
nulls_percent_df[nulls_percent_df['nulls_percentage']!=0]
````

````python
# Looking at the DB I can see that there are tons of empty cells that have some value in them, maybe a space
cats
````

````python
# Yes, it's definitely a space so I need to remove that, probably all across the DF 
cats['MAILCODE'].value_counts()
````

````python
cats['MAILCODE'] = cats['MAILCODE'].apply(lambda x: x.replace(" ", "A"))
````

````python
cats = cats.apply(lambda x: x.replace(" ", np.NaN))
cats
````

````python
# Now that looks a bit more like it 
nulls_percent_df = pd.DataFrame(cats.isna().sum()/len(cats)*100).reset_index()
nulls_percent_df
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df
nulls_percent_df[nulls_percent_df['nulls_percentage']!=0]
````

````python
# list of columns to drop, including the two specified in the description
columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>50]
columns_above_threshold['column_name']
drop_columns_list = list(columns_above_threshold['column_name'])
drop_columns_list += ['OSOURCE']
drop_columns_list += ['ZIP']
print(drop_columns_list)
````

````python
# Now a cute function to remove all of those in the list from the DF
def drop(df, columns_to_drop) :
    return df.drop(columns=columns_to_drop)
````

````python
clean = drop(cats,drop_columns_list)
clean
````

Perform all of the cleaning processes from the Lesson.

Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

Note that there are a few null values in the column. We will first replace those null values using the code below:
print(categorical['GENDER'].value_counts())

````python
df = pd.DataFrame(clean['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
clean['STATE'] = list(map(clean_state, clean['STATE']))
````

````python
# This looks a bit dodgy, a mix of strings and values. I'll replace the strings 
clean['NOEXCH'].value_counts()
````

````python
# Yay
clean['NOEXCH'] = clean['NOEXCH'].replace("0", 0)
clean['NOEXCH'] = clean['NOEXCH'].replace("1", 1)
````

````python
# Need to remove the NaNs and replace them with the median

mode_values = clean.mode().iloc[0]  # Get the first row of the mode DataFrame

# Replace NaN values with the mode for each column
clean_full = clean.fillna(mode_values)
````

````python
clean_full
````

````python
# Going to keep these columns because we did it all in the lab 
final = clean_full[['STATE',	'MDMAUD',	'CLUSTER',	'HOMEOWNR',	'GENDER',	'DATASRCE',	'RFA_2R',	'RFA_2A',	'GEOCODE2']]
final
````

````python
print(final['GENDER'].value_counts())
final['GENDER'] = final['GENDER'].fillna('F')
````

````python
final['GENDER'] = final['GENDER'].apply(lambda x: x.replace("A", "U"))
final['GENDER'] = final['GENDER'].apply(lambda x: x.replace("C", "U"))
final['GENDER'] = final['GENDER'].apply(lambda x: x.replace("J", "U"))
final['GENDER'] = final['GENDER'].apply(lambda x: x.replace("U", "other"))
final['GENDER'].value_counts()
````
